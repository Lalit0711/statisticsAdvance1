Q1. What is a random variable in probability theory?

- In probability theory, a random variable is a function that assigns a numerical value to each outcome of a random experiment. It helps us quantify uncertain outcomes in a mathematical way. For example, when tossing a coin, we can assign 1 to heads and 0 to tails—this assignment is done by a random variable.
There are two main types of random variables: discrete and continuous. A discrete random variable takes on a countable number of values, such as the number of heads in three coin tosses. A continuous random variable, on the other hand, can take any value within a range or interval, such as the time it takes to complete a task.

Q2. What are the types of random variables?

 1. Discrete Random Variable -

 A discrete random variable is one that takes on a finite or countable number of distinct values. These values are often integers and result from experiments with specific, separate outcomes. For example, when rolling a die, the result can be any one of the six values: 1, 2, 3, 4, 5, or 6. Similarly, the number of heads in three coin tosses is a discrete value such as 0, 1, 2, or 3. Discrete random variables are associated with a Probability Mass Function (PMF), which gives the probability of each possible value. Each individual outcome has a non-zero probability, and the sum of all probabilities equals 1.

 2. Continuous Random Variable -

 A continuous random variable can take on infinitely many values within a given interval on the real number line. Unlike discrete variables, the exact value of a continuous variable is not counted but measured, such as time, temperature, height, or weight. For instance, the time taken to finish a race might be 12.31 or 12.3156 seconds, and any value within a range is possible. Continuous random variables are described using a Probability Density Function (PDF). The probability of the variable taking an exact value is zero; instead, we calculate the probability over a range of values using integration.

Q3. What is the difference between discrete and continuous distributions?

1. Discrete Distribution:

* Deals with countable outcomes.
* Values can be finite or countably infinite (e.g., 0, 1, 2, ...).
* Uses Probability Mass Function (PMF).
* Probability is assigned to specific values.
* Example: Number of heads in coin tosses, number of students in a class.
* $P(X = x)$ is meaningful and non-zero.
* Total probability: $\sum P(X = x_i) = 1$.



2. Continuous Distribution:

* Deals with uncountable outcomes.
* Values lie in a continuous range (e.g., any value between 0 and 1).
* Uses Probability Density Function (PDF).
* Probability is assigned to intervals, not exact values.
* Example: Height, weight, time, temperature.
* $P(X = x) = 0$; only $P(a \leq X \leq b)$ makes sense.
* Total probability: $\int_{-\infty}^{\infty} f(x) \, dx = 1$.


Q4.  What are probability distribution functions (PDF)?

- A Probability Distribution Function (PDF) describes how the values of a random variable are distributed. For a **discrete random variable**, the PDF (also called the probability mass function) gives the probability that the variable takes each specific value. For a continuous random variable, the PDF is a function that represents the relative likelihood of the variable falling within a certain range of values — although the probability at an exact point is zero, the area under the curve over an interval gives the probability for that interval. PDFs are essential tools in statistics and probability theory for understanding data behavior.


Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF))

- Key Differences:

 PDF gives the probability density at a point (or exact probabilities for discrete cases), while
CDF gives the total probability up to a point.

 For continuous variables, the PDF is the derivative of the CDF, and the CDF is the integral of the PDF.

 CDF is always non-decreasing and ranges from 0 to 1.
PDF values can be more than 1, but the total area under the PDF curve equals 1.


Q6. What is a discrete uniform distribution?

 - A discrete uniform distribution is a type of probability distribution where all possible outcomes are equally likely. It applies to a finite set of values, and each value has the same probability of occurring. For example, when rolling a fair six-sided die, each number from 1 to 6 has a 1/6 chance, making it a discrete uniform distribution.

Q7.  What are the key properties of a Bernoulli distribution?

- The Bernoulli distribution is a simple probability distribution for a random experiment with only two outcomes: success (1) and failure (0). Here are its key properties:

1. Two outcomes: Only 0 and 1 (e.g., win/lose, yes/no).
2. One parameter: Defined by a single parameter p, the probability of success (where $0 \leq p \leq 1$).
3. Mean (Expected value): $E(X) = p$
4. Variance: $\text{Var}(X) = p(1 - p)$
5. Probability mass function (PMF):

   $$
   P(X = x) =
   \begin{cases}
   p & \text{if } x = 1 \\
   1 - p & \text{if } x = 0
   \end{cases}
   $$

 It’s commonly used to model **single yes/no trials** like flipping a biased coin.

Q8. What is the binomial distribution, and how is it used in probability?

- The binomial distribution is a probability distribution that shows the likelihood of getting a certain number of successes in a fixed number of independent trials, where each trial has only two outcomes (success or failure) and the same probability of success.

 For example, if you flip a coin 10 times and want to know the chance of getting exactly 6 heads, you can use the binomial distribution.

 It is defined by two parameters:

* n = number of trials
* p = probability of success in each trial

Q9. What is the Poisson distribution and where is it applied?

- The Poisson distribution is a probability distribution used to model the number of times an event occurs in a fixed interval** of time, space, or area, when the events happen independently and at a constant average rate.

 It is defined by a single parameter λ(lambda), which represents the average number of occurrences in the given interval.

 Applications:

* Number of customer arrivals at a shop per hour
* Number of emails received in a day
* Number of phone calls at a call center
* Number of accidents at an intersection per week

Q10. What is a continuous uniform distribution?

- A continuous uniform distribution is a probability distribution where all values in a given interval are equally likely to occur. It is defined over a continuous range from a to b, and the probability density is constant throughout this interval.

 Key features:

* Every value between **a and b** has the **same probability density**.
* The PDF is:

  $$
  f(x) = \frac{1}{b - a} \quad \text{for } a \leq x \leq b
  $$
* The mean is $\frac{a + b}{2}$, and the variance is $\frac{(b - a)^2}{12}$.

Q11. What are the characteristics of a normal distribution?

- The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution with the following key characteristics:

1. Bell-shaped curve: It is symmetric around the mean.
2. Mean = Median = Mode: All are equal and located at the center.
3. Symmetry: The left and right sides of the curve are mirror images.
4. Defined by two parameters:

   * Mean (μ): determines the center
   * Standard deviation (σ): determines the spread
5. 68-95-99.7 Rule:

   * About 68% of the data falls within 1σ of the mean
   * 95% within 2σ
   * 99.7% within 3σ
6. Total area under the curve is 1, representing the total probability.
7. Tails extend infinitely in both directions, getting closer to the axis but never touching it.

Q12. What is the standard normal distribution, and why is it important?

- The standard normal distribution is a special case of the normal distribution where the mean is 0 and the standard deviation is 1. It follows the same bell-shaped curve but is centered at 0 and has a consistent spread.

 It is important because:

* It simplifies calculations in statistics, especially for probabilities and hypothesis testing.
* Any normal distribution can be converted to the standard normal using a process called z-transformation:

  $$
  z = \frac{x - \mu}{\sigma}
  $$
* Statistical tables (z-tables) are based on the standard normal and help find areas (probabilities) under the curve.

Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

- The central limit theorem (CLT) is a key concept in statistics that says when you take many random samples from any population and calculate their averages, the distribution of those averages will be approximately normal (bell-shaped), even if the original population is not normal, as long as the sample size is large enough.

 It is important because it allows us to use normal distribution methods to make predictions, build confidence intervals, and perform hypothesis testing — even when the data doesn't come from a normal population. This makes the CLT a powerful tool in real-world statistics.


Q14. How does the Central Limit Theorem relate to the normal distribution?

- The central limit theorem relates to the normal distribution by explaining why the normal distribution appears so often in statistics. It states that if you take many random samples from any population and calculate their means, the distribution of those sample means will tend to follow a normal distribution, especially as the sample size increases.

 This happens regardless of the shape of the original population. Because of this, the normal distribution becomes a useful tool for analyzing averages and making predictions, even when the original data is not normally distributed.

Q15. What is the application of Z statistics in hypothesis testing?

- Z statistics is used in hypothesis testing to determine whether there is a significant difference between sample data and a population parameter when the population standard deviation is known.

 In practice, you calculate a **Z-score** to measure how far your sample statistic is from the population mean in terms of standard deviations. This score is then compared to a critical value from the standard normal (Z) distribution to decide whether to **reject or fail to reject the null hypothesis**.

 Common applications include:

* Testing if a sample mean differs from a known population mean
* Comparing two means or proportions when sample sizes are large
* Constructing confidence intervals

Q16.  How do you calculate a Z-score, and what does it represent?


- $$
Z = \frac{X - \mu}{\sigma}
$$

 Where:

* X is the value from the dataset
* μ is the population mean
* σ is the population standard deviation

 The Z-score tells you how many standard deviations a value X is from the mean.

 What it represents:

* A Z-score of 0 means the value is exactly at the mean.
* A positive Z-score means the value is above the mean.
* A negative Z-score means the value is below the mean.

Q17. What are point estimates and interval estimates in statistics?

- In statistics, point estimates and interval estimates are two ways to estimate a population parameter based on sample data.

* Point estimate is a single value that serves as the best guess of an unknown population parameter. For example, the sample mean is a point estimate of the population mean.

* Interval estimate gives a range of values within which the population parameter is likely to lie, along with a certain level of confidence. This is called a confidence interval. For example, a 95% confidence interval might say the true mean lies between 48 and 52.

 While point estimates are simple, interval estimates provide more information by showing the uncertainty in the estimate.

Q18.  What is the significance of confidence intervals in statistical analysis?


- Confidence intervals are important in statistical analysis because they give a range of values within which the true population parameter is likely to fall, along with a confidence level (like 95% or 99%).

 The significance of confidence intervals includes:

1. Measuring uncertainty: They show how precise your estimate is, rather than just giving a single value.
2. Decision making: They help determine if a parameter is within an acceptable range for business, science, or policy decisions.
3. Hypothesis testing: If the confidence interval does not include a value (like 0 or a target mean), it can suggest statistical significance.
4. More informative than point estimates: They provide context and reliability around your sample statistics.

Q19. What is the relationship between a Z-score and a confidence interval?

- The Z-score and confidence interval are closely related in statistics because the Z-score helps determine the range (margin of error) for a confidence interval when the population standard deviation is known.

 In a confidence interval, the Z-score corresponds to the desired confidence level. For example:

* For a 90% confidence level, Z ≈ 1.645
* For a 95% confidence level, Z ≈ 1.96
* For a 99% confidence level, Z ≈ 2.576

 The formula for a confidence interval using Z-score is:

$$
\text{Confidence Interval} = \bar{x} \pm Z \times \left( \frac{\sigma}{\sqrt{n}} \right)
$$

Q20. How are Z-scores used to compare different distributions?

- Z-scores help compare values from different distributions by showing how far each value is from its average, using the same scale.

 Even if two sets of data have different averages and spreads, Z-scores convert the values so you can compare them fairly. For example, if you scored 80 in one test and 75 in another, Z-scores can tell you which score was better compared to others in that test.

 So, Z-scores are like a common measuring stick to compare results from different situations.

Q21. What are the assumptions for applying the Central Limit Theorem?

- The Central Limit Theorem (CLT) has a few key assumptions to apply correctly:

1. Random sampling: The data should be collected using random sampling methods.
2. Independence: Each observation should be independent of the others.
3. Sample size: The sample size should be large (usually **n ≥ 30**). Smaller samples may work if the population is already normally distributed.
4. Finite variance: The population should have a finite mean and variance.

 When these conditions are met, the CLT ensures that the sampling distribution of the sample mean will be approximately normal, even if the original population is not.

Q22. What is the concept of expected value in a probability distribution?

- The expected value in a probability distribution is the average or mean outcome you would expect if an experiment or process were repeated many times.

 It is calculated by multiplying each possible value of the random variable by its probability and then adding all those results together.

 Formula:

 For a discrete random variable:

$$
E(X) = \sum [x \cdot P(x)]
$$

Q23.  How does a probability distribution relate to the expected outcome of a random variable?

- A probability distribution shows all the possible values a random variable can take and the likelihood of each value. The expected outcome, or expected value, is calculated using this distribution and represents the average result over many repetitions of the random process.

 In simple terms, the probability distribution provides the weights (probabilities) for each possible value, and the expected value is the weighted average of those values. It tells us what we can expect to happen on average in the long run, based on the probabilities given by the distribution.

```python
# Q3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):

    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1")

    return p**x * (1 - p)**(1 - x)

print("P(X=0):", bernoulli_pdf(0, 0.7))  # Output: 0.3
print("P(X=1):", bernoulli_pdf(1, 0.7))  # Output: 0.7
```

```python
#Q9. Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):

    mean = np.mean(data)
    std = np.std(data)

    z_scores = (data - mean) / std

    plt.hist(z_scores, bins=30, edgecolor='black', density=True, alpha=0.7)
    plt.title('Z-score Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    return z_scores

data = np.random.normal(50, 10, 1000)
z_scores = calculate_and_plot_zscores(data)
```

```python
# Q15. Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm

sample = [52, 49, 50, 51, 48, 53, 47, 49, 50, 51]
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)

mu = 50

z = (sample_mean - mu) / (sample_std / np.sqrt(n))
print(f"Z-statistic: {z:.2f}")

alpha = 0.05
z_critical = norm.ppf(1 - alpha / 2)
print(f"Critical Z-value: ±{z_critical:.2f}")

if abs(z) > z_critical:
    print("Reject the null hypothesis (H₀)")
else:
    print("Fail to reject the null hypothesis (H₀)")
```

```python
# Q16. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy import stats

data = [82, 85, 88, 90, 86, 87, 84, 83, 89, 91]

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)
standard_error = sample_std / np.sqrt(n)

confidence = 0.95
z_critical = stats.norm.ppf((1 + confidence) / 2)

margin_of_error = z_critical * standard_error
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
# Q17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
from scipy import stats

np.random.seed(42)
mu, sigma = 100, 15
sample_size = 50
data = np.random.normal(mu, sigma, sample_size)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
standard_error = sample_std / np.sqrt(sample_size)

confidence = 0.95
z_critical = stats.norm.ppf((1 + confidence) / 2)
margin_of_error = z_critical * standard_error

ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval for Mean: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# Q21. Write a Python program to compare the standard deviations of two datasets and visualize the difference.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
data1 = np.random.normal(loc=50, scale=5, size=1000)
data2 = np.random.normal(loc=50, scale=15, size=1000)

std1 = np.std(data1, ddof=1)
std2 = np.std(data2, ddof=1)

print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")

plt.figure(figsize=(10, 5))
plt.hist(data1, bins=30, alpha=0.6, label=f'Dataset 1 (std={std1:.2f})', color='skyblue')
plt.hist(data2, bins=30, alpha=0.6, label=f'Dataset 2 (std={std2:.2f})', color='salmon')
plt.title("Comparison of Standard Deviations")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Q22.  Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np

np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=1000)

data_range = np.max(data) - np.min(data)

q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print(f"Range: {data_range:.2f}")
print(f"Q1 (25th percentile): {q1:.2f}")
print(f"Q3 (75th percentile): {q3:.2f}")
print(f"IQR (Q3 - Q1): {iqr:.2f}")
```

```python
# Q23.  Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
data = np.random.normal(loc=100, scale=20, size=1000)

mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Original Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, color='salmon', edgecolor='black')
plt.title("Z-score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Q24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_normal_distribution_skew_kurtosis(mean=0, std=1, size=1000):
    data = np.random.normal(loc=mean, scale=std, size=size)

    skewness = skew(data)
    kurt = kurtosis(data, fisher=False)

    print(f"Sample Mean: {np.mean(data):.2f}")
    print(f"Sample Std Dev: {np.std(data, ddof=1):.2f}")
    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis: {kurt:.4f}")

analyze_normal_distribution_skew_kurtosis()
```
